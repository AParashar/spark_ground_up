```python
from pyspark.sql import SparkSession
import getpass
username = getpass.getuser()

spark = SparkSession. \
builder. \
config('spark.ui.port','0'). \
config("spark.sql.warehouse.dir", f"/user/{username}/warehouse"). \
enableHiveSupport(). \
master('yarn'). \
getOrCreate()
```

```python
sc = spark.sparkContext
```

```python
covid_cases_rdd = sc.textFile("data/datasets/covid19/cases/covid_dataset_cases.csv")
covid_states_rdd = sc.textFile("data/datasets/covid19/states/covid_dataset_states.csv")
```

## 3. Top 15 States having maximum no.of recovery

```python
cases_col = covid_cases_rdd.map(lambda x: (x.split(",")[1], int(x.split(",")[11])))
```

```python
cases_col_agg = cases_col.reduceByKey(lambda x,y: x+y).sortBy(lambda x: x[1], ascending = False)
```

```python
cases_col_agg.take(10)
```
