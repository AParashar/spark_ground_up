```python
from pyspark.sql import SparkSession
import getpass
username = getpass.getuser()

spark = SparkSession. \
builder. \
config('spark.ui.port','0'). \
config("spark.sql.warehouse.dir", f"/user/{username}/warehouse"). \
enableHiveSupport(). \
master('yarn'). \
getOrCreate()
```

```python
sc = spark.sparkContext
```

```python
orders_base_rdd = sc.textFile("data/retail_db_orders_part-00000")
customers_base_rdd = sc.textFile("data/retail_db_customers_part-00000")
order_items_base_rdd = sc.textFile("data/retail_db_order_items_part-00000")
```

## 4. Top 3 states with Maximum customers

```python
customer_states = customers_base_rdd.map(lambda x: (x.split(",")[7], 1))
```

```python
customer_states.take(10)
```

```python
# State, Customer_id
customer_states.reduceByKey(lambda x,y: x+y).sortBy(lambda x: x[1], False).take(10)
```

```python
customer_count_states = customer_states.countByValue()
```
